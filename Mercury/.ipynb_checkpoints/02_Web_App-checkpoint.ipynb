# Historic stock price üìà

```python
ticker = mr.Select(value="TSLA", 
                   choices=["TSLA", "COIN", "MSFT", "PLTR", "SNOW", "NFLX"], 
                   label="Select stock")
```

```python
history = yf.Ticker(ticker.value).history(period="3mo")
```

mr.Md(f"""## Basic statistics
